Нормализация базы данных

База данных уже находится в 1НФ — все атрибуты в таблице простые, нет повторений строк, все используемые домены содержат только скалярные значения.
Чтобы привести базу данных ко 2НФ, создаем отдельные таблицы address и product. (Если дробить еще сильнее, можно было бы также выделить отдельную таблицу order с полями online_order, order_status, customer_id, product_id, а также таблицу wealth_segment c полями job_title, job_industry_category, wealth_segment, owns_car, property_valuation, customer_id). Теперь каждый столбец, который не является ключом, зависит от первичного ключа.
После преобразования таблица находится в 3НФ — нет транзитивных связей.
